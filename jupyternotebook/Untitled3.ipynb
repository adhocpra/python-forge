```python
class World_cup:
    def __init__(self,matches,winner,venue,prize,year):
        self.matches=matches
        self.winner=winner
        self.venue=venue
        self.prize=prize
        self.year=year
    
    def show_me(self):
        print(f"Matches: {self.matches},Winner:{self.winner},Venue:{self.venue},Prize:{self.prize}, Year:{self.year}")

matches = int(input("Enter number of matches :"))
winner= input("Enter the winner :")
venue=input("Enter Venue :")
prize=input("Enter prizes :")
year=int(input("enter thr playing year :"))

this_year= World_cup (matches,winner,venue,prize,year)
this_year.show_me()


matches = int(input("Enter number of matches :"))
winner= input("Enter the winner :")
venue=input("Enter Venue :")
prize=input("Enter prizes :")
year=int(input("enter thr playing year :"))

next_year=World_cup (matches,winner,venue,prize,year)
next_year.show_me()
```
